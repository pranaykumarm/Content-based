```python
# For randomforest, all features must be numbers
def normalizeMimeType(mimeType):
	if mimeType == 'text/html':
		return 1
	elif mimeType == 'text/plain':
		return 2
	elif mimeType == 'application/javascript':
		return 3
	elif mimeType == 'application/x-javascript':
		return 4
	elif mimeType == 'application/json':
		return 5
	elif mimeType == 'image/jpeg':
		return 6
	elif mimeType == 'image/png':
		return 7
	elif mimeType == 'image/gif':
		return 8
	elif mimeType == 'text/css':
		return 9
	else:
		return 0

def normalizeRequestMethod(method):
	if method == 'POST':
		return 1
	elif method == 'GET':
		return 2
	elif method == 'CONNECT':
		return 3
	elif method == 'HEAD':
		return 4
	elif method == 'PUT':
		return 5
	else:
		return 0

def normalizeUrlCounts(csvfile):
	# Grab all URLs in file
	f = open(csvfile, 'rU')
	headers = ['timestamp','processtime','sourceIP','responseMethod','responseSize','requestMethod','URL','rfc931','peerStatusHost','mimeType']
	logfile = csv.DictReader(f, fieldnames=headers, dialect='excel')
	urlList = [{}]
	urlPerDomain = Counter()
	# Build a list of all the URLs
	for row in logfile:
		parsed = urlparse(row['URL'])
		(domain,uri) = (parsed[1], parsed[2])
		a = {domain: uri}
		if a not in urlList:
			urlList.append(a)

	urlPerDomain = Counter(chain.from_iterable(domain.keys() for domain in urlList))

	return urlPerDomain
```

```python
def parse(csvfile, label):
# Grab features from CSV file and put them into an array
	f = open(csvfile, 'rU')
	# Squid logs don't have headers, so we need to define them
	headers = ['timestamp','processtime','sourceIP','responseMethod','responseSize','requestMethod','URL','rfc931','peerStatusHost','mimeType']
	logfile = csv.DictReader(f, fieldnames=headers, dialect='excel')
	labels = []
	featuresArray = []
	# Build the normalized URL counts
	urlPerDomain = normalizeUrlCounts(csvfile)
	prevrow = ""
	for row in logfile:
		# Convert all log entries to interger equivalents (normalization)
		# Get time deltas
		if prevrow == "":
			timeDelta = 0
		else:
			timeDelta = row['timestamp'] - prevrow['timestamp']
		# Normalize MIME type and method
		mimeType = normalizeMimeType(row['mimeType'])
		method = normalizeRequestMethod(row['requestMethod'])

		if row['responseSize'] >= 500:
			isLarge = 1
		else:
			isLarge = 0
		# Get the URL per domain count for the current domain
		domain = urlparse(row['URL'])[1]
		URLcount = urlPerDomain[domain]

		# Build array from normalized data features
		featuresArray.append([timeDelta,isLarge,URLcount,method,mimeType])
		labels.append(label)
		# Save previous row
		row = prevrow

	f.close()
	return featuresArray, labels
```
