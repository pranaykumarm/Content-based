## Training set

```python
K = len(app_set)+len(eid_set)
K
```

```python
a = len(app_set)
a
```

```python
user_act = {}
for name in aun_set:
    user_act[name] = []

for i in xrange(n):
    vec = np.array([0 for k in xrange(K)])
    vect_ind = 0
    for e in app_set:
        if app[i]==e:
            vec[vect_ind] = 1
            break
        vect_ind += 1
    
    vect_ind = 0
    for e in eid_set:
        if eid[i]==e:
            vec[a+vect_ind] = 1
            break
        vect_ind += 1
            
    user_act[aun[i]].append(vec)
```

```python
train = []

for i in xrange(10000):
    vec = np.array([0 for k in xrange(K)])
    vect_ind = 0
    for e in app_set:
        if app[i]==e:
            vec[vect_ind] = 1
            break
        vect_ind += 1
    
    vect_ind = 0
    for e in eid_set:
        if eid[i]==e:
            vec[a+vect_ind] = 1
            break
            
    train.append(vec)
```

## OCSVM

```python
from scipy.cluster.vq import vq, kmeans, whiten
```

```python
from sklearn import svm
```

```python
C_range = 10.0 ** np.arange(-8, 0)
gamma_range = 10.0 ** np.arange(-8, 1)
param_grid = dict(gamma=gamma_range, nu=C_range)
```

```python
len(train)
```

```python
best_nu = 0
best_gamma = 0
best_sco = 0

for nu in C_range:
    for gamma in gamma_range:
        clf = svm.OneClassSVM(nu=nu, kernel="rbf", verbose=True, gamma=gamma)
        clf.fit(train)
        sco = clf.predict(train)
        if np.sum(sco) >= best_sco :
            best_nu = nu
            best_gamma = gamma
            best_sco = np.sum(sco)
        del clf
```

```python
def score(events, avg):
    
    norm = np.linalg.norm(avg)
    sco = []

    for event in events:
        sco.append((np.dot(event,avg)*1.0)/(norm*math.sqrt(2)))
        
    return sco
```
